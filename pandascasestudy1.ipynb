```python
duplicated = df.duplicated().sum()
print("Duplicated Values are:",duplicated)
if duplicated == 0:
  print("No Need to remove Any Duplicates")
else:
  df.drop_duplicates(inplace=True)
```

```python
#Rename columns ("Title" > "Movie_Title")
newdf = df.rename(columns = {'Title':'Movie_Title'})
newdf
```

```python
#Get unique genres

genres = newdf['Genre'].str.split(',').explode().unique()
genres
```

```python
newdf
```

```python
#Extract number of genres per movie

newdf['Genre_Count'] = newdf['Genre'].str.split(",").str.len()
print(newdf[['Movie_Title', 'Genre_Count']])
```

```python
#Handle missing values in 'Revenue (Millions)' and 'Metascore
print(newdf[['Revenue (Millions)','Metascore']].isnull().sum())
newdf['Revenue (Millions)'] = newdf['Revenue (Millions)'].fillna(
    newdf['Revenue (Millions)'].mean()
)

newdf['Metascore'] = newdf['Metascore'].fillna(
    newdf['Metascore'].mean()
)
newdf
```

```python
#Top 5 movies by Rating

top5 = newdf.sort_values(by='Rating', ascending=False).head(5)
top5
```

```python
#Movies with highest/lowest revenue

# Movies with highest revenue
print("Movie With Highest Revenue:")
print(
    newdf[newdf['Revenue (Millions)'] ==
          newdf['Revenue (Millions)'].max()]
    [['Movie_Title', 'Revenue (Millions)']]
)

print()

# Movies with lowest revenue
print("Movie With Lowest Revenue:")
print(
    newdf[newdf['Revenue (Millions)'] ==
          newdf['Revenue (Millions)'].min()]
    [['Movie_Title', 'Revenue (Millions)']]
)
```

```python
#Avg. rating by Director

avg_rating = newdf.groupby('Director')['Rating'].mean()
avg_rating
```

```python
#Movies per year

movies_per_year = newdf['Year'].value_counts()
movies_per_year
```

```python
#Movies with rating > 8 & revenue > 100M

newrating = newdf[(newdf['Rating'] > 8 ) & (newdf['Revenue (Millions)'] > 100)].head()
newrating
```

```python
#Group by Director: avg Rating, Revenue

director_group = newdf.groupby('Director')[['Rating','Revenue (Millions)']].mean()
director_group
```

```python
#Group by Year: count, avg Rating

year_stats = newdf.groupby('Year').agg({
    'Movie_Title': 'count',
    'Rating': 'mean'
}).round(2)

print(year_stats)
```

```python
#Create a new column Profitability
#Revenue divided by Duration
#Find top 10 most profitable movies

print("Revenue Divided By Duration")
print()
newdf['Profitability'] = newdf['Revenue (Millions)'] / newdf['Runtime (Minutes)']

print(newdf[['Movie_Title', 'Profitability']].head())
print()
print("TOP 10 MOVIES")
top10 = newdf.sort_values(
    by='Profitability',
    ascending=False
)

print(top10[['Movie_Title', 'Profitability']].head(10))
```

```python
#Create a column that categorizes movies by rating
#Rating < 6: Poor
#6 ≤ Rating < 7.5: Average
#Rating ≥ 7.5: Excellent

newdf['Rating_Category'] = newdf['Rating'].apply(
    lambda x: 'Poor' if x < 6
    else 'Average' if x < 7.5
    else 'Excellent'
)

print(newdf[['Movie_Title', 'Rating', 'Rating_Category']].head())
```

```python
#Create a custom function to score movies based on:
#Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
#Apply a function to dataset to assign the score

def score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

newdf['Score'] = newdf.apply(score, axis=1)

print(newdf[['Movie_Title', 'Score']].head())
```
